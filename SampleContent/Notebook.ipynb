```python
import requests
from config import app_key
import pandas as pd
```

```python
ticker = 'GOOG'
endpoint = f'https://api.tdameritrade.com/v1/marketdata/{ticker}/pricehistory'

payload = {
    'apikey': app_key,
    'periodType': 'day',
    'period': '1',
    'frequencyType': 'minute',
    'frequency': '1',
    'needExtendedHoursData': 'false'
}

content = requests.get(url = endpoint, params = payload)

data = content.json()
data = pd.DataFrame(data['candles'])
data
```

```python
tickers = 'GOOG,TSLA,MSFT'
endpoint = 'https://api.tdameritrade.com/v1/marketdata/quotes'

payload = {
    'apikey': app_key,
    'symbol': tickers
}

content = requests.get(url = endpoint, params = payload)

data = content.json()
data
```
